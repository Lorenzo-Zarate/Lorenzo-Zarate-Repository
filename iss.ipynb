```python
import requests
import time
import matplotlib.pyplot as plt
```

```python
latitude = []
longitude = []
timestamps = []

for i in range(100):
    response = requests.get("https://api.wheretheiss.at/v1/satellites/25544")
    data = response.json()
    lat = data['latitude']
    long = data['longitude']
    timest = data['timestamp']

    latitude.append(lat)
    longitude.append(long)
    timestamps.append(timest)
    
    print(f"Point {i}: {data['latitude']}, {data['longitude']}, {data['timestamp']}")
    
    time.sleep(10)

plt.title("Tracking the ISS")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.plot(longitude, latitude, marker='o', linestyle='-', color='blue')
plt.show()
```
